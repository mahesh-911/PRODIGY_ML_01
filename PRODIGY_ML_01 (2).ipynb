```python
# I am not using train_test_split function as I am using the entire data for training and the entire data for testing as well
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#unzipping the file
import zipfile
zip_r = zipfile.ZipFile('/content/boston_housing.zip')
zip_r.extractall('/content')
zip_r.close()
```

```python
data = pd.read_csv('/content/BostonHousing.csv')
y_data = data['medv']
y_test = data['medv']
print(y_data)
print(y_test)
```

```python
#just printing the first five rows of the dataset
data.head()
```

```python
x_train = data
x_train.drop(['medv'],axis=1,inplace = True)
print(x_train)
```

```python
x_test = x_train
print(x_test)
print(x_train)
```

```python
data.head()
#checking are there any null values in the dataset
data.isnull().sum()
```

```python
#printing the independent variables and their values
x_data = data
print(x_data)
```

```python
#just describing the data like count, mean,standard deviation...
x_data.describe()
```

```python
#using the linear regression model
model = LinearRegression()
model.fit(x_train,y_data)
```

```python
#predicting the values, printing the predicted values and the actual values to compare them
y_pred = model.predict(x_test)
print(y_pred)

print(y_test)
```

```python
# checking the accuracy using mean squared error
mse = mean_squared_error(y_pred,y_test)
print("Mean_Squared_error: ",mse)
```
